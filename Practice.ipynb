```python
#pseudo-log-return
def pseufoLogReturn(dataFrame, closingColumn, interval):
    column = dataFrame[closingColumn]
    rowNum = dataFrame.shape[0]
    averages = []
    for i in range(int(np.floor(rowNum/interval))):
        data = column.iloc[i*interval:(i+1)*interval]
        averages.append(np.mean(data))
    if not np.isnan(np.mean(column[(i+1)*interval:])):
        averages.append(np.mean(column[(i+1)*interval:]))
    
    result = []
    for i in range(1,len(averages)):
        temp = np.log(averages[i]/averages[i-1])*100
        result.append(temp)
    return result

#print(pseufoLogReturn(selected, "Adj. Close", 5))

plt.hist(pseufoLogReturn(selected, "Adj. Close", 5),bins=50)
plt.show()
```

```python
#log return
def logReturn(dataFrame, closingColumn, interval):
    column = dataFrame[closingColumn]
    rowNum = dataFrame.shape[0]
    closingPrices = []
    for i in range(interval-1,rowNum,interval):
        closingPrices.append(column[i])
    if i!=(rowNum-1):
        closingPrices.append(column[-1])
    
    result = []
    for i in range(1,len(closingPrices)):
        temp = np.log(closingPrices[i]/closingPrices[i-1])*100
        result.append(temp)
    return result

#logReturn = logReturn(selected, "Adj. Close", 5)
plt.hist(logReturn(selected, "Adj. Close", 5),bins=100)
plt.show()
```

```python
#5-day trend
def trend(dataFrame, closingColumn, interval):
    column = dataFrame[closingColumn]
    rowNum = dataFrame.shape[0]
    coeff = []
    for i in range(int(np.floor(rowNum/interval))):
        data = column.iloc[i*interval:(i+1)*interval]
        x = np.linspace(1,interval,interval)
        coeff.append(tuple(np.polyfit(x,data,1)))
    if not column[(i+1)*interval:].empty:
        x = np.linspace(1,rowNum-(i+1)*interval,rowNum-(i+1)*interval)
        data = column[(i+1)*interval:]
        coeff.append(tuple(np.polyfit(x,data,1)))
    return coeff
trend(selected, "Adj. Close", 5)
```

```python
#create table that groups adj closing prices for each trend group (5 days) and gives their average and standard deviationn

def df_test(dataFrame, price_column, interval):
    weeks = list(range(0,52))
    averages = []
    stdevs = []
    prices = []
    for j in range(len(dataFrame)):
        prices.append(dataFrame.iloc[j][price_column])
    for i in range(1,len(prices)+1):
        if i%interval == 0:
            average = np.mean(prices[i-interval:i-1])
            averages.append(average)
            deviation = np.std(prices[i-interval:i-1])
            stdevs.append(deviation)
    data = list(zip(weeks, averages, stdevs))
    df = pd.DataFrame(data, columns = ['Week', 'Avg Adj Close', 'St Dev'])
    return df

test = df_test(selected, 'Adj. Close', 5)
test
```

```python
import math
def init(inputNum, layerNum, outputNum):
    result = []
    for i in range(layerNum):
        if i == 0:
            print("0 to 1st")
            result.append(np.zeros((inputNum,inputNum)))
        if i>0:
            size = (math.ceil((inputNum-i+1)*inputNum/layerNum), math.ceil((inputNum-i)*inputNum/layerNum))
            result.append(np.zeros(size))
        if i == layerNum-1:
            size = (math.ceil((inputNum-i)*inputNum/layerNum), outputNum)
            result.append(np.zeros(size))
    return result
```

```python
def mean_sq_error(data1, data2):
    sq_differences = []
    for i in range(len(data1)):
        actual = data1[i]
        predicted = data2[i]
        sq_difference = (actual - predicted)**2
        sq_differences.append(sq_difference)
    MSE = 1/(i+1)*sum(sq_differences)
    return MSE
```

```python
def mean_abs_percent_error(data1, data2):
    abs_differences = []
    for i in range(len(data1)):
        actual = data1[i]
        predicted = data2[i]
        abs_difference = abs(actual - predicted)/abs(actual)
        abs_differences.append(abs_difference)
    MAPE = 100/(i+1)*sum(abs_differences)
    return MAPE
```

```python
def direction_acc(data1, data2):
    products = []
    for i in range(len(data1)):
        actual = data1[i]
        predicted = data2[i]
        product = actual*predicted
        products.append(product)
    DA = 100/(i+1)*sum(products)
    return DA
```

```python
#i think we should just go with bidirectional by default; shorten the function after running each case to see

#length is number of consecutive time steps ('sequence of features') -- using the dataset
#test, where we group by 5 days, this gives us 50 timesteps
#our goal is to output value of next time step -- so the next week's average

##dropout seems to be something used with RNN and the rate can be adjusted
##not entirely sure what our units would be; the article uses 256
#num_layers = 7 default b/c 1 in, 1 out, 5 hidden
def create_model(length = 50, units = 250, cell = LSTM, num_layers = 7, dropout = 0.3,
                loss = 'mean_squared_error', optimizer = 'adam', 
                 bidirectional = False):
    model = Sequential()
    for i in range(num_layers):
        #input layer
        if i == 0:    
            if bidirectional:
                model.add(Bidirectional(cell(units, return_sequences=True), input_shape=(None, length)))
            else:
                model.add(cell(units, return_sequences=True, input_shape=(None, length)))
        #output layer
        elif i == num_layers-1:
            if bidirectional:
                model.add(Bidirectional(cell(units, return_sequences=False)))
            else:
                model.add(cell(units, return_sequences=False))
        #hidden layers
        else:
            if bidirectional:
                model.add(Bidirectional(cell(units, return_sequences=True)))
            else:
                model.add(cell(units, return_sequences=True))
        
        model.add(Dropout(dropout))
        
    model.add(Dense(1,activation = 'linear'))
    model.compile(loss=loss, metrics = ['mean_squared_error'], optimizer=optimizer)
    return model

model = create_model()
```
